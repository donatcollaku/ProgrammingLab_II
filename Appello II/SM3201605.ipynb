## Esercizio 1

Crea un vettore NumPy con le seguenti spese annuali per l’auto (in euro) sostenute da 7 persone:
3200, 2750, 4100, 3600, 2950, 5100, 4400

1. Calcola la spesa totale complessiva sostenuta dal gruppo.

2. Le persone con spese minori di 3000 € ottengono una detrazione fiscale del 15%. Applica la modifica al vettore.

3. Quanto ha risparmiato il gruppo dopo la detrazione fiscale?


````python
import numpy as np
import pandas as pd

#Ho inserito nell'array i valori dati dall'esercizio
a = np.array([3200, 2750, 4100, 3600, 2950, 5100, 4400], dtype=float)

#1
#Uso la funzione sum() per fare la somma di tutti i valori presenti nell'array
spesaTotale = a.sum()
print(f"La spesa totale sostenuta dal gruppo equivale a: {spesaTotale} Euro")

#2
print(f"Prima della detrazione:")
print(f"Spese minori di 3.000 Euro: {a[a < 3000]} ")
#Tra le parentesi quadre inserisco la condizione per la quale mi viene tornato un array di Boolean che assegna True esclusivamente alla celle che rispettano la condizione "a < 3000"
#A tutti i valori che rispecchiano questa condizione vengono detratti 15%
a[a < 3000] *= 0.85
print(f"Post detrazione:")
print(f"Spese minori di 3.000 Euro: {a[a < 3000]} ")

#3
#Qua andiamo a fare la differenza tra le spese totali prima e dopo le detrazioni
risparmio = np.abs(a.sum() - spesaTotale)
print(f"Il gruppo ha risparmiato: {risparmio} Euro ({spesaTotale} - {a.sum()})")



````

## Esercizio 4 

Utilizza il dataset cereal.csv.

Contiene dati nutrizionali (proteine, fibre, calorie, zuccheri, ecc.) per circa 70 cereali.


1.  Aggiungi una colonna chiamata `nutri_score`, definita come:
$$
\texttt{nutri\_score} = \frac{\texttt{protein} + \texttt{fiber}}{\texttt{calories}}
$$

Indica quanti nutrienti positivi ci sono per ogni caloria: più è alto, più il cereale è "nutriente a parità di energia.

2. Trova i **5 cereali con il `nutri_score` più alto** e stampali con `name` e `nutri_score`.

3. Filtra i cereali che soddisfano entrambi i criteri:

   - meno di 120 calorie  
   - più di 2 grammi di fibra 


4. **Crea un grafico a barre** dei **10 cereali con il `nutri_score` più alto**, usando `name` sull’asse x.

5. Crea uno **scatter plot** (grafico a punti) che visualizzi:

   - `calories` sull’asse x  
   - `nutri_score` sull’asse y
   - `protein` nel colore dei punti


````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("cereal.csv", sep=";")

print(df.head(10))

#1
#Aggiungo una nuova colonna
df["nutri_score"] = ((df["protein"] + df["fiber"]) / df["calories"])
print(f"File: {df.head(10)}")

#2
#Ordino il dataframe per i valori nutriscore e seleziono solo i 5 valori piu alti, di cui andremo a fare l'output solo delle colonne name e nutri_score
print(f"Cereali con il nutri score maggiore: {df[["name", "nutri_score"]].sort_values(by="nutri_score", ascending=False).head(5)}")

#3
#Vado a creare la condizione per la quale filtro i valori del dataframe e compio l'output
condizione = (df["calories"] < 120) & (df["fiber"] > 2)
print(f"Cereali con meno di 120 calorie e piu di 2 grammi di fibre: {df[condizione]}")

#4 Grafico
valoriGrafico = df.sort_values(by="nutri_score", ascending=False).head(10)
valoriGrafico

plt.title("10 cereali con Nutri Score maggiore")
plt.xlabel("Nome cereali")
plt.ylabel("Nutri Score")
plt.bar(valoriGrafico["name"], valoriGrafico["nutri_score"])
plt.show()


#5
plt.title("Rapporto Calorie - Nutri Score")
plt.xlabel("Calorie")
plt.ylabel("Nutri Score")
plt.scatter(df["calories"], df["nutri_score"])
plt.show()


````

$$

L’**intervallo di confidenza al 95%** si approssima con:

$$
\text{parametro}_i \pm 1.96 \times \text{errore\_standard}_i
$$
